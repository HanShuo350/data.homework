```python
class MySet:
    def __init__(self):
        self._data = []
    """实例化"""
    def insert(self, element):
        if element not in self._data:
            self._data.append(element)
            return True
        return False

    def delete(self, element):
        if element in self._data:
            self._data.remove(element)
            return True
        return False

    def query(self, element):
        return element in self._data

if __name__ == '__main__':
    s = MySet()
    fruits = ['apple', 'banana', 'cherry', 'orange', 'mango', 'pear']
    for f in fruits:
        s.insert(f)
    print(s.query('apple'))
    s.delete('banana')
    print(s.query('banana'))
```

```python
def insert_sorted(s, value):
    """
    在升序有序数组 arr 中插入 value，保持数组有序。
    """
    left, right = 0, len(s)
    while left < right:
        mid = (left + right) // 2
        if s[mid] < value:
            left = mid + 1
        else:
            right = mid
    pos = left
    s.insert(pos, value)
    return s

sorted_list = [1, 3, 5, 7, 9]
insert_sorted(sorted_list, 4)
print(sorted_list)
```
