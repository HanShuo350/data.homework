```python
fruits: list[str] = ['apple', 'banana', 'cherry','orange','mango']
```

```python
from typing import Iterable,Any
class DeleteItem:
    def __init__(self, iterable: list[Any]):
        """初始化时传入待操作的列表"""
        self.iterable = iterable
    def delete(self,index_to_delete:int) -> list[Any]:
        """
        删除列表中指定索引的元素
        :param index_to_delete:删除元素的索引
        """
        length_of_iterable: int = len(self.iterable)
        for i in range(index_to_delete, len(self.iterable)-1):
            self.iterable[i] = self.iterable[i + 1]

        self.iterable.pop()

        return self.iterable




```

```python
DeleteItem(fruits).delete(3)
print(fruits)
```
