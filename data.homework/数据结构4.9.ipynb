```python
import random

def quick_sort_random(arr, left=0, right=None):
    """随机化快速排序"""
    if right is None:
        right = len(arr) - 1

    if left < right:
        # 随机选择轴并交换到最后
        pivot_idx = random.randint(left, right)
        arr[pivot_idx], arr[right] = arr[right], arr[pivot_idx]

        # 分区
        pivot = arr[right]
        i = left - 1

        for j in range(left, right):
            if arr[j] <= pivot:
                i += 1
                arr[i], arr[j] = arr[j], arr[i]

        arr[i + 1], arr[right] = arr[right], arr[i + 1]
        pivot_pos = i + 1

        # 递归排序左右两部分
        quick_sort_random(arr, left, pivot_pos - 1)
        quick_sort_random(arr, pivot_pos + 1, right)

    return arr
```

```python
def quick_sort_3way(arr, left=0, right=None):
    """
    三路快速排序
    适用于包含大量重复元素的数组
    """
    if right is None:
        right = len(arr) - 1

    if left >= right:
        return arr

    # 随机选择轴点（优化）
    pivot_idx = random.randint(left, right)
    arr[pivot_idx], arr[left] = arr[left], arr[pivot_idx]
    pivot = arr[left]

    # 三路分区指针
    # lt: 小于区域的右边界
    # i: 当前扫描位置
    # gt: 大于区域的左边界
    lt = left      # arr[left+1...lt] < pivot
    i = left + 1   # arr[lt+1...i-1] == pivot
    gt = right     # arr[gt...right] > pivot

    while i <= gt:
        if arr[i] < pivot:
            arr[lt], arr[i] = arr[i], arr[lt]
            lt += 1
            i += 1
        elif arr[i] > pivot:
            arr[i], arr[gt] = arr[gt], arr[i]
            gt -= 1
        else:  # arr[i] == pivot
            i += 1

    # 递归排序小于和大于的部分
    quick_sort_3way(arr, left, lt - 1)
    quick_sort_3way(arr, gt + 1, right)

    return arr
```

```python
# 测试随机化快速排序
print("=" * 50)
print("1. 随机化快速排序测试")
arr1 = [3, 6, 8, 10, 1, 2, 1]
print(f"原数组: {arr1}")
quick_sort_random(arr1)
print(f"排序后: {arr1}")

# 测试已排序数组（最坏情况）
arr2 = [1, 2, 3, 4, 5, 6, 7]
print(f"\n已排序数组: {arr2}")
quick_sort_random(arr2)
print(f"排序后: {arr2}")

# 测试三路快速排序
print("\n" + "=" * 50)
print("2. 三路快速排序测试")
arr3 = [3, 1, 2, 3, 1, 2, 3, 1, 2]
print(f"原数组: {arr3}")
quick_sort_3way(arr3)
print(f"排序后: {arr3}")

# 测试大量重复元素
arr4 = [2, 2, 1, 1, 3, 3, 2, 1, 3, 2]
print(f"\n大量重复元素: {arr4}")
quick_sort_3way(arr4)
print(f"排序后: {arr4}")

# 性能对比（可选）
print("\n" + "=" * 50)
print("3. 性能对比（大量重复元素）")
import time

# 生成测试数据
test_arr = [random.randint(1, 5) for _ in range(10000)]

# 随机化快排
arr_copy1 = test_arr.copy()
start = time.time()
quick_sort_random(arr_copy1)
time1 = time.time() - start

# 三路快排
arr_copy2 = test_arr.copy()
start = time.time()
quick_sort_3way(arr_copy2)
time2 = time.time() - start

print(f"随机化快速排序: {time1:.4f} 秒")
print(f"三路快速排序: {time2:.4f} 秒")
print(f"三路快排提升: {(time1/time2 - 1)*100:.1f}%")
```
